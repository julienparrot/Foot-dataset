```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
dataset = pd.read_csv("C:\\Users\\julie\\OneDrive\\Bureau\\m1\\s\\SUPPORTS\\data\\foot.csv ", header=0)
dataset.info()
```

```python
dataset
```

```python
plt.figure(figsize=(12,7))
sns.countplot(dataset.Goals)
plt.title('Moyenne du nomber de but mis par les joueurs')
```

```python
dataset.drop(dataset.index[(dataset["Goals"] <= 11)],axis=0,inplace=True)
```

```python
sns.lmplot(x="Mins", y="Goals", data=dataset)
plt.title('Nuage de points qui représente les buts par rapport au minute joué')
```

```python
sns.distplot(dataset.Mins, bins=20)
plt.title('Moyenne du nombre total de minute jouées par les joueurs')
```

```python
dataset.drop(dataset.index[(dataset["Mins"] <= 3000)],axis=0,inplace=True)
```

```python
dataset
```

```python
sns.lmplot(x="Shots", y="OnTarget", data=dataset)
plt.title('Nuage de points qui représente les tirs cadrés par rapport au total de tir')
```

```python
sns.countplot(dataset.Country)
plt.title('Histogramme représentant le nombre de joueur par pays du club')
```

```python
plt.figure(figsize=(20,7))
sns.countplot(dataset.League)
plt.title('Histogramme représentant le nombre de joueur par ligue')
```

```python
dataset.sort_values(by='Goals')
```

```python
dataset.drop(dataset.index[(dataset["Goals"] <= 30)],axis=0,inplace=True)
```

```python
dataset.sort_values(by='Goals')
```

```python
dataset.drop(dataset.index[(dataset["League"] == 'MLS')],axis=0,inplace=True)
```

```python
dataset.sort_values(by='Goals')
```

```python
matrix = dataset.corr() 
print(matrix)
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(matrix, vmin=0, vmax=1, annot=True)
```

```python
plt.figure(figsize=(16, 6))
sns.boxplot(x="Matches_Played", y="Goals", data=dataset, linewidth=2.5)
plt.title('Boxplot représentant la moyenne de but par rapport au nombre de matche joué')
```

```python
price = dataset[['Player_Names', 'Goals', 'Year']].groupby('Player_Names').median().round(2)
fig, axes = plt.subplots(figsize=(15,5), dpi=100)
plt.bar(price.index, height=price.Goals)
plt.title('Record but dans une saison');
```
